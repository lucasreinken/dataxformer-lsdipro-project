```python
from dotenv import load_dotenv
import os

load_dotenv()

config = {
    "host": os.getenv("VERTICA_HOST"),
    "port": os.getenv("VERTICA_PORT"),
    "user": os.getenv("VERTICA_USER"),
    "password": os.getenv("VERTICA_PASSWORD"),
    "database": os.getenv("VERTICA_DATABASE"),
    "tlsmode": "disable",
}
```

```python
from vertica_python import connect

conn = connect(**config)

cur = conn.cursor()
cur.execute("""
    SELECT table_schema, table_name
    FROM tables
    ORDER BY table_schema, table_name;
""")
for row in cur.fetchall():
    print(row)

print("columns_tokenized")
cur.execute("""
    SELECT
        column_name,
        data_type,
        ordinal_position
    FROM columns
    WHERE table_name = 'columns_tokenized'
    ORDER BY ordinal_position;
""")


for row in cur.fetchall():
    print(row)

cur.execute("SELECT COUNT(*) FROM public.webtable_AllTables;")
count = cur.fetchone()[0]
print("Row count:", count)

print("inverse_document_frequency")
cur.execute("""
    SELECT
        column_name,
        data_type,
        ordinal_position
    FROM columns
    WHERE table_name = 'inverse_document_frequency'
    ORDER BY ordinal_position;
""")

for row in cur.fetchall():
    print(row)

print("columns_tokenized")
cur.execute("""
    SELECT
        column_name,
        data_type,
        ordinal_position
    FROM columns
    WHERE table_name = 'columns_tokenized'
    ORDER BY ordinal_position;
""")

for row in cur.fetchall():
    print(row)

print("queries")
cur.execute("""
    SELECT
        column_name,
        data_type,
        ordinal_position
    FROM columns
    WHERE table_name = 'queries'
    ORDER BY ordinal_position;
""")

for row in cur.fetchall():
    print(row)

print("webtable_AllTables")
cur.execute("""
    SELECT
        column_name,
        data_type,
        ordinal_position
    FROM columns
    WHERE table_name = 'webtable_AllTables'
    ORDER BY ordinal_position;
""")

print("yagoFacts")
cur.execute("""
    SELECT
        column_name,
        data_type,
        ordinal_position
    FROM columns
    WHERE table_name = 'yagoFacts'
    ORDER BY ordinal_position;
""")

for row in cur.fetchall():
    print(row)

cur.execute("""
    SELECT TableId
    FROM webtable_AllTables
    LIMIT 1000;
""")
rows = cur.fetchall()
for r in rows:
    print(r)
conn.close()
```
