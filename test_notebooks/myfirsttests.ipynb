```python
def create_projections(table_list, 
                       stemmer, 
                       stop_words: list = None, 
                       given_stem_cache: dict = None, 
                       return_stem_cache:bool = False, 
                       min_word_len = 0, 
                       create_ngrams:bool = False, 
                       ngrams_size:int = 2
                       ) -> dict: 

    ps = stemmer
    
    stemmed_stopwords = set()
    if stop_words: 
        for stop_word in stop_words: 
            stemmed_stopword = ps.stem(stop_word)
            stemmed_stopwords.add(stemmed_stopword)

    if given_stem_cache: 
        stem_cache = given_stem_cache.copy()
    else: 
        stem_cache = dict() 

    projections = dict() 

    for table_index, table in enumerate(table_list): 
        for row_index, row in enumerate(table): 
            for column_index, cell in enumerate(row): 

                               ####Das hier als Helper für Indexing? 

                ngram_stem_cache = []

                for word in words: 
                    if word in stem_cache: 
                        stemmed_word = stem_cache.get(word)
                    else: 
                        stemmed_word = ps.stem(word)
                        stem_cache[word] = stemmed_word
                        
                    if stemmed_word not in stemmed_stopwords and len(stemmed_word) >= min_word_len: 

                        projections.setdefault(stemmed_word, set()).add((table_index, row_index, column_index))

                        if create_ngrams: 
                            ngram_stem_cache.append(stemmed_word)
                
                if create_ngrams: 
                    created_ngrams = ngrams(ngram_stem_cache, ngrams_size)
                    ######Das hier könnten auch single Strings und keine Tuple sein. Also ein .join(_) vielleicht?
                    for created_ngram in created_ngrams: 
                        projections.setdefault(created_ngram, set()).add((table_index, row_index, column_index))

    if return_stem_cache: 
        return projections, stem_cache
    
    return projections
```

```python
def cleaner(value, create_ngrams:bool = False, ngram_size:int = 2): 
    sentence = value.lower().translate(str.maketrans("", "", string.punctuation))
    words = word_tokenize(sentence)
    stemmed_words = []
    for word in words: 
        stemmed_word = ps.stem(word)
        stemmed_words.append(stemmed_word)

    if create_ngrams: 
        created_ngrams = ngrams(stemmed_words, ngram_size)
        stemmed_words.extend(list(created_ngrams))
    return stemmed_words
```

```python
def indexing(cleaned_values:dict, projections:dict)->dict: 
    index_dict = dict() 
    for value in cleaned_values: 
        value_index = projections.get(value, None)

        if value_index: 
            for index_pair in value_index: 
                table_id, row_id, col_id = index_pair 
                index_dict.setdefault((table_id, row_id), set()).add((col_id))
    
    return index_dict
```

```python
def getAnswers(Querries: set, tables:dict, table_list:list) -> dict:

    """
    Returns a Dict of all the Answers to all Querries for all relevant Tables of a Search 
    In: 
        Querries: cleaned set of Querries for wich we need the Answers
        Tables: Dict key: Table_ID -> (Row_ID, (ColX_ID, ColY_ID)) of all the Relevent Tables
        Table_List: A List of all Tables 
    
    Out: 
        Answers: A Dict of Answers Key: x -> Value: Set(y_1, ..., y_n)
        If len(answers[x])==1, no contradictions exist 
    """
    answers = dict()
    for table_id, tup in tables.items(): 
        row_id, col_tup = tup
        x_col_id, y_col_id = col_tup 
        table = table_list.get(table_id)
        for row in table: 
            x_in_question = row[x_col_id]
            for x in Querries: 
                if x == x_in_question: 
                    answers.setdefault(x, set()).add(cleaner(row[y_col_id])) 
    
    return answers

```
